```python
import pandas as pd
import numpy as np

df = pd.read_csv("data/data.csv")
df.head()
```

```python
 NUMS = ["impressions","clicks","spent","total_conversion","approved_conversion"]

# parsing dates
for d in ["reporting_start","reporting_end"]:
    df[d] = pd.to_datetime(df[d],errors="coerce",dayfirst = True)

# parse numerics
for c in NUMS:
    df[c] = pd.to_numeric(df[c], errors="coerce")

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=["total_conversion","approved_conversion"])
df.isnull().sum()
```

```python
# validation
df.describe().T
```

```python
df = df[df["clicks"] <= df["impressions"]]
```

```python
# date parsing
df["day"] = df["reporting_start"].dt.date
df.head()
```

```python
# building a segment key
df["segment_key"] = (df["fb_campaign_id"]+"|" +
                     df["age"].astype(str) + "|" +
                     df["gender"].astype(str) + "|" +
                     df["interest1"].astype(str)
)
df.head()
```

## Computing KPIs
--CTR = clicks / impressions

--CVR = approved_conversion / clicks

--CPC = spent / clicks

--CPA = spent / approved_conversion

--ROAS = (AOV × approved_conversion) / spent (AOV = average order value you’ll set in the app)
These match Google’s ROAS definition (“conversion value / cost”

```python
# computing KPIs
AOV = 80.0
g = (df.groupby(["day","segment_key"],as_index = False).agg(
    impressions = ("impressions","sum"),
    clicks = ("clicks","sum"),
    spent = ("spent","sum"),
    conv = ("approved_conversion","sum"))
)
g.head()
```

```python
g.shape
```

```python
# avoid divide-by-zero with clip
g["ctr"]  = g["clicks"] / g["impressions"].clip(lower=1)
g["cvr"]  = g["conv"]   / g["clicks"].clip(lower=1)
g["cpc"]  = g["spent"]  / g["clicks"].clip(lower=1)
g["cpa"]  = g["spent"]  / g["conv"].clip(lower=1)
g["roas"] = (AOV * g["conv"]) / g["spent"].clip(lower=1e-9)
```

```python
g.to_parquet("data/facts_daily.parquet")
```
